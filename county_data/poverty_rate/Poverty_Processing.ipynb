# Auxilary Functions

```python
def move_to_processed(filename):
    current_path = '~/school/spring2020/cse4940/datasets/poverty_rate/raw/'
    destination_path = 'processed/'
    
    os.rename(filename, destination_path+filename)

# Updates Counties InPlace
def format_counties(dataframe):
    for i in range(len(dataframe)):
        county = str(dataframe.loc[i, 'County FIPS'])
        if len(county) < 3:
            if len(county) == 2:
                dataframe.loc[i, 'County FIPS'] = '0' + county
            else:
                dataframe.loc[i, 'County FIPS'] = '00' + county
    
def format_states(dataframe):
    for i in range(len(dataframe)):
        state = str(dataframe.loc[i, 'County FIPS'])
        if len(state) < 2:
            dataframe.loc[i, 'County FIPS'] = '0' + state
    
def process_poverty(poverty):
    
    poverty = poverty.copy()
    
    # 1. Get Columns
    columns = []
    for col in poverty.loc[0, :]:
        columns.append(col)

    # 2. Drop first three rows
    # 3. Reset Columns
    poverty.columns = columns
    poverty = poverty.drop([0]).reset_index().drop(['index'], axis=1)

    poverty = poverty[['State FIPS', 'County FIPS', 
                           'Poverty Percent All Ages', 'Median Household Income']]
    
    # Ensure Counties and States are formatted properly
    format_counties(poverty)
    format_states(poverty)
    
    # Get True County FIPS
    for i in range(len(poverty)):
        prefix = str(poverty.loc[i, 'State FIPS'])
        suffix = str(poverty.loc[i, 'County FIPS'])
        poverty.loc[i, 'County FIPS'] = prefix + suffix
        
    poverty = poverty.drop(['State FIPS'], axis=1)
    
    return poverty
```

# Load Files

```python
poverty2009 = pd.read_excel('raw/est09all.xls')
poverty2010 = pd.read_excel('raw/est10all.xls')
poverty2011 = pd.read_excel('raw/est11all.xls')
poverty2012 = pd.read_excel('raw/est12all.xls')
poverty2013 = pd.read_excel('raw/est13all.xls')
poverty2014 = pd.read_excel('raw/est14all.xls')
poverty2015 = pd.read_excel('raw/est15all.xls')
poverty2016 = pd.read_excel('raw/est16all.xls')
```

# Process Files

```python
poverty2009 = process_poverty(poverty2009)
poverty2010 = process_poverty(poverty2010)
poverty2011 = process_poverty(poverty2011)
poverty2012 = process_poverty(poverty2012)
poverty2013 = process_poverty(poverty2013)
poverty2014 = process_poverty(poverty2014)
poverty2015 = process_poverty(poverty2015)
poverty2016 = process_poverty(poverty2016)
```
