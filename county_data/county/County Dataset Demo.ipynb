```python
import pandas as pd
```

# Load County Datasets

```python
county2009 = pd.read_csv('county2009_final.csv').drop(['Unnamed: 0'], axis=1)
county2010 = pd.read_csv('county2010_final.csv').drop(['Unnamed: 0'], axis=1)
county2011 = pd.read_csv('county2011_final.csv').drop(['Unnamed: 0'], axis=1)
county2012 = pd.read_csv('county2012_final.csv').drop(['Unnamed: 0'], axis=1)
county2013 = pd.read_csv('county2013_final.csv').drop(['Unnamed: 0'], axis=1)
county2014 = pd.read_csv('county2014_final.csv').drop(['Unnamed: 0'], axis=1)
county2015 = pd.read_csv('county2015_final.csv').drop(['Unnamed: 0'], axis=1)
county2016 = pd.read_csv('county2016_final.csv').drop(['Unnamed: 0'], axis=1)
county2017 = pd.read_csv('county2017_final.csv').drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
county2018 = pd.read_csv('county2018_final.csv').drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
```
